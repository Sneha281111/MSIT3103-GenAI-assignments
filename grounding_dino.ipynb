```python
import os
from PIL import Image

# Set the file you actually have:
image_path = "cat.jpg"   # or "kitty.jpg" if that's your file

# If you uploaded a .webp, convert once:
if image_path.endswith(".webp"):
    Image.open(image_path).convert("RGB").save("cat.jpg", "JPEG")
    image_path = "cat.jpg"

print("Using:", image_path, " | Exists:", os.path.exists(image_path))
```

```python
from PIL import Image
img = Image.open(image_path).convert("RGB")
max_side = 1024  # try 768 if you want it faster
if max(img.size) > max_side:
    scale = max(img.size) / max_side
    new_wh = (int(img.width/scale), int(img.height/scale))
    img = img.resize(new_wh)
    img.save("resized.jpg", "JPEG", quality=92)
    image_path = "resized.jpg"
print("Final image:", image_path)
```

```python
from groundingdino.util.inference import load_image, predict, annotate
import matplotlib.pyplot as plt

image_source, image = load_image(image_path)

TEXT_PROMPT    = "cat . shiny eyes . fluffy ears . pink nose . whiskers"
BOX_THRESHOLD  = 0.40     # raise to reduce false positives
TEXT_THRESHOLD = 0.28

boxes, logits, phrases = predict(
    model=model,
    image=image,
    caption=TEXT_PROMPT,
    box_threshold=BOX_THRESHOLD,
    text_threshold=TEXT_THRESHOLD,
    device="cpu"     # important on CPU runtime
)

annotated = annotate(image_source=image_source, boxes=boxes, logits=logits, phrases=phrases)

plt.figure(figsize=(10,8))
plt.imshow(annotated)
plt.axis("off")
plt.show()
```

```python
import cv2   # <-- add this

prompts = [
    "cat . shiny eyes . fluffy ears . pink nose . whiskers",
    "kitten . paw . tail . collar",
    "sleeping cat . pillow . blanket"
]

for i, p in enumerate(prompts, 1):
    with torch.inference_mode():
        boxes, logits, phrases = predict(
            model=model,
            image=image,
            caption=p,
            box_threshold=0.40,
            text_threshold=0.28,
            device="cpu"
        )
    ann = annotate(image_source=image_source, boxes=boxes, logits=logits, phrases=phrases)
    cv2.imwrite(f"cat_detected_{i}.jpg", ann[..., ::-1])   # save result

print("Saved: cat_detected_1.jpg, cat_detected_2.jpg, cat_detected_3.jpg")
```
